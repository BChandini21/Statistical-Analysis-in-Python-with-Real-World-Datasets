<b> Sample Code for Statistical Analysis in Python: </b>

<b> 1. Loading the Breast_cancer Dataset: </b>

<b> Import Libraries: </b>

pandas is for handling data.

load_breast_cancer loads a breast cancer dataset.

<b> Load Data: </b>

The breast_cancer dataset is loaded, containing tumor features and whether they are benign or malignant.

<b> Create a DataFrame:</b>

The data is organized into a table (DataFrame) with feature names as columns, and a new "target" column is added for the diagnosis (0 = benign, 1 = malignant).

<b> Display Data:</b>

It prints the first few rows to preview the data.

```python
import pandas as pd
from sklearn.datasets import load_breast_cancer

# Load the dataset
breast_cancer = load_breast_cancer()
df = pd.DataFrame(data=breast_cancer.data, columns=breast_cancer.feature_names)
df['target'] = breast_cancer.target

# Display the first few rows
print(df.head())
```

<b> 2. Performing Descriptive Statistics: </b>

<b> 1. Mean, Median, Mode:</b> These measure the central tendency of the dataset.

df.mean(): The average value for each column.

df.median(): The middle value.

df.mode(): The most frequent value.

<b> 2. Standard Deviation and Variance:</b> These measure data spread.

df.std(): Measures how much data deviates from the mean.

df.var(): The square of the standard deviation.

<b> 3. Range, Skewness, Kurtosis:</b>
df.max() - df.min(): The difference between maximum and minimum values (range).

df.skew(): Asymmetry of the data.

df.kurt(): How heavy the tails of the distribution are.

```python
# Calculate basic descriptive statistics
print("Mean:\n", df.mean())
print("\nMedian:\n", df.median())
print("\nMode:\n", df.mode().iloc[0])
print("\nStandard Deviation:\n", df.std())
print("\nVariance:\n", df.var())

# Additional descriptive statistics
print("\nRange:\n", df.max() - df.min())
print("\nSkewness:\n", df.skew())
print("\nKurtosis:\n", df.kurt())
```

<b> 3. Performing Inferential Statistics: </b>

<b> 1. One-Sample T-Test:</b> This is a statistical test used to determine if the average (mean) of your sample data is different from a known value (the hypothetical mean).

<b> 2. Hypothetical Mean:</b> This is the value you want to compare your sample mean against. For example, if you think the average mean radius should be 14.0, you use that as your hypothetical mean.

<b> 3. T-Statistic:</b> This number tells you how much your sample mean differs from the hypothetical mean, considering the variability in your data. A larger absolute t-statistic indicates a bigger difference.

<b> 4. P-Value:</b> This is the probability of seeing a result as extreme as yours, or more extreme, if the hypothetical mean is actually true. A small p-value suggests that the observed difference is unlikely to have happened by chance.

<b> 5. Significance Level (Alpha):</b> This is a threshold you set (commonly 0.05). If the p-value is smaller than this threshold, it suggests that the observed difference is statistically significant.

<b> Interpretation:</b>

if the p-value is less than 0.05, you conclude that the average value of your feature is likely different from the hypothetical mean.

If the p-value is greater than 0.05, the evidence isn’t strong enough to say that your sample mean is different from the hypothetical mean.

```python
from scipy import stats

# Select the feature of interest, for example, 'mean radius'
feature_values = df['mean radius']

# Hypothetical population mean to compare against
hypothetical_mean = 14.0  # This is an example value; you can change it as needed

# Perform a one-sample t-test
t_stat, p_value = stats.ttest_1samp(feature_values, hypothetical_mean)

# Print the results
print("T-Statistic:", t_stat)
print("P-Value:", p_value)

# Interpretation
alpha = 0.05  # Significance level (commonly used value)

if p_value < alpha:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

<b> 4. Confidence Intervals: </b>

<b> 1. Find the Average Value:</b>

Calculate the average mean radius from your data.

<b> 2. Estimate How Much This Average Could Vary:</b>

Determine how much this average might be off from the true average.

<b> 3. Create Range Around the Average:</b>

Use the average and the estimated variation to make a range. This range is where you expect the true average to be 95% of the time.

<b> Show the Range:</b>

Print this range. It’s like saying, "I’m 95% sure the real average is somewhere in this range."

```python
import numpy as np

mean_radius_values = df['mean radius']

# Sample mean and standard error for the selected feature (mean radius)
sample_mean = np.mean(mean_radius_values)
standard_error = stats.sem(mean_radius_values)

# Compute the 95% confidence interval for the mean radius
confidence_interval = stats.norm.interval(0.95, loc=sample_mean, scale=standard_error)

print("95% Confidence Interval for Mean Radius:", confidence_interval)
```

<b> 5. Regression Analysis:</b>

<b> 1. import statsmodels:</b>

This library is used for statistical modeling.

<b> 2. Prepare the Data: </b>

X = sm.add_constant(df['mean radius']): Adds a constant term to the independent variable mean radius. This constant represents the intercept in the regression model. You can replace 'mean radius' with any other feature you want to analyze.

y = breast_cancer.target: Defines the dependent variable. This is the target variable you are trying to predict (e.g., labels for cancer diagnosis).

<b> 3. Fit the Linear Regression Model:</b>

model = sm.OLS(y, X).fit(): Fits a linear regression model using Ordinary Least Squares (OLS) where y is the dependent variable and X is the independent variable(s) with the constant term.

<b> 4. Print the Model Summary:</b>

print(model.summary()): Displays a detailed summary of the regression results, including coefficients, R-squared value, p-values, and other statistics.

```python
import statsmodels.api as sm

# Define independent variable (add constant for intercept)
X = sm.add_constant(df['mean radius'])  # You can replace 'mean radius' with any other feature from the Breast Cancer dataset

# Define dependent variable (e.g., target variable in the Breast Cancer dataset)
y = breast_cancer.target

# Fit linear regression model
model = sm.OLS(y, X).fit()

# Print model summary
print(model.summary())
```

<b> 6. Interpreting Regression Model Results: Coefficients, P-Values, and R-Squared: </b>

<b> 1. Extract Values:</b>

Coefficients: These are the numbers that show the impact of each feature on the target variable.

P-Values: These help you determine if the coefficients are statistically significant.

R-Squared: This tells you how well your model explains the variability in the target variable.

<b> 2. Print Results:</b>

Display the coefficients, p-values, and R-squared value.

<b> 3. Interpret Results:</b>

P-Values: If a p-value is less than 0.05, it means that the corresponding coefficient is statistically significant (likely to be a real effect). If it's greater than 0.05, it’s not considered significant.

R-Squared: Shows how much of the target variable’s variability is explained by your model. For example, if R-squared is 0.80, your model explains 80% of the variability.

```python
# Extracting specific values
coefficients = model.params
p_values = model.pvalues
r_squared = model.rsquared

# Interpretation
print("\nInterpretation")
print("Coefficients:")
print(coefficients)
print("\nP-Values:")
print(p_values)
print("\nR-Squared:")
print("R-Squared: ", r_squared)

# Basic interpretation of p-values
print("\nInterpretation of P-Values:")
print("If the p-value is less than 0.05, the coefficient is considered statistically significant. Otherwise, it is not.")

# R-Squared interpretation
print("\nR-Squared:", round(r_squared, 2))
print("The model explains", round(r_squared * 100, 2), "% of the variability in the target variable.")
```

<b> 7. Visualizing the Relationship Between Mean Radius and Tumor Classification with Regression Line: </b>

<b> 1. Prepare Data:</b> 

Get your feature (mean radius) and target (e.g., cancer type).

<b> 2. Fit the Model:</b>

Create a model to predict the target using the feature.

<b> 3. Get Residuals:</b>

Find the differences between the actual values and what the model predicted.

<b> 4. Plot Residuals:</b>

Create a histogram to see how these differences are spread out.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define the independent variable and dependent variable
X = sm.add_constant(df['mean radius'])  # Adding constant for intercept
y = breast_cancer.target

# Fit linear regression model
model = sm.OLS(y, X).fit()

# Create a DataFrame for predictions
df['predicted'] = model.predict(X)

# Plot the scatter plot and regression line
sns.scatterplot(x=df['mean radius'], y=y, label='Actual Data', color='pink')
sns.lineplot(x=df['mean radius'], y=df['predicted'], label='Regression Line', color='orange')
plt.xlabel('Mean Radius')
plt.ylabel('Target')
plt.title('Relationship Between Mean Radius and Target with Regression Line')
plt.gcf().set_facecolor('black') 
plt.gca().set_facecolor('black') 
plt.legend()
plt.show()
```
